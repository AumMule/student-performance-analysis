# NumPy Operations on Marks

```python
marks_array = np.asarray(df["marks"])
mean_marks = np.mean(marks_array)
median = np.median(marks_array)
sd = np.std(marks_array)
print(f"Median:{median}\nStandard Deviation:{sd}\nMean:{mean_marks}")
```

## Assigning grades to the rows based on marks

```python
grades = np.select(
    [df["marks"]>=95,df["marks"]>=90,df["marks"]>=85,df["marks"]>=80,df["marks"]>=75],
    ["A","B","C","D","E"],
    "F"
)
df["grades"] = grades
```
